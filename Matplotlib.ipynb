```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'r-')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('T')
#plt.show() for printing if not automatic
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('T')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.9,.9])
axes2=fig.add_axes([.6,.1,.4,.3]) #left, bottom, width, height
axes1.plot(x,y)
axes1.set_title('T1')
axes2.plot(y,x)
axes2.set_title('T2')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=3) #can't handle mutiple rows
axes[0].plot(x,y)
axes[0].set_title('T1')
axes[1].plot(y,x)
axes[1].set_title('T2')

#for c in axes:
#    c[0].plot(x,y)
#    c[1].plot(y,x)

#    for i in c:
#        i.plot(x,y)

#axes.plot(x,y)
plt.tight_layout()
```

```python
fig=plt.figure(figsize=(6,4))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(6,4))
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(6,4))
axes[0].plot(x,y,label='x vs. y')
axes[0].plot(y,x,label='y vs. x')
axes[0].legend(loc=0) #0=best, can also use tuple (0.1,0.1)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#FF8C00', lw=4, alpha=0.4, ls='-.',marker='^',
        markersize=10, markerfacecolor='blue',
        markeredgewidth=3, markeredgecolor='red') #RGB hex code or 'orange' #'steps'
ax.set_xlim([0,4])
ax.set_ylim([0,5])
```

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,.2,.2])
ax1.plot(x,y,'r')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax2.plot(x,y,'r')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
```

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,.4,.4])
ax1.plot(x,z,'r')
ax1.set_xlabel('X')
ax1.set_ylabel('Z')
ax2.plot(x,y,'r')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
```

```python
fig,ax=plt.subplots(nrows=1, ncols=2,figsize=(12,2))
ax[0].plot(x,y,'b',lw=3,ls='--')
ax[1].plot(x,z,'r',lw=3)
```
