```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
dt=make_blobs(n_samples=200, n_features=2,centers=4,cluster_std=1.8,random_state=101)
```

```python
dt[1]
```

```python
plt.scatter(dt[0][:,0],dt[0][:,1],c=dt[1])
```

```python
from sklearn.cluster import KMeans #unsupervised
```

```python
km=KMeans(n_clusters=4)
```

```python
km.fit(dt[0])
```

```python
km.cluster_centers_
```

```python
km.labels_
```

```python
fig,(ax1,ax2)=plt.subplots(1,2,sharey=True,figsize=(12,6))
ax1.set_title('K Means')
ax1.scatter(dt[0][:,0],dt[0][:,1],c=km.labels_,cmap='rainbow')
ax2.set_title('Original')
ax2.scatter(dt[0][:,0],dt[0][:,1],c=dt[1],cmap='rainbow')
```
